```python
from typing import TypedDict
from langgraph.graph import StateGraph, START, END
```

```python
class BatsmanState(TypedDict):
    runs: int
    balls: int
    fours: int
    sixes: int
    sr: float
    bpb: float
    boundary_percentage: float
    summary: str
```

```python
def calculate_sr(state: BatsmanState):
    sr = (state["runs"] / state["balls"]) * 100
    return {"sr": sr}
```

```python
def calculate_bpb(state: BatsmanState):
    bpb = (state["fours"] + state["sixes"]) / state["balls"]
    return {"bpb": bpb}
 
```

```python
def calculate_boundary_percentage(state: BatsmanState):
    boundary_percentage = (state["fours"] * 4 + state["sixes"] * 6) / state["runs"] * 100
    return {"boundary_percentage": boundary_percentage}
```

```python
def summary(state: BatsmanState):
    print(f"Runs: {state['runs']}")
    print(f"Balls: {state['balls']}")
    print(f"Fours: {state['fours']}")
    print(f"Sixes: {state['sixes']}")
    print(f"Strike Rate: {state['sr']:.2f}")
    print(f"Balls Per Boundary: {state['bpb']:.2f}")
    print(f"Boundary Percentage: {state['boundary_percentage']:.2f}%")
    return {"summary": "done"}
```

```python
graph = StateGraph(BatsmanState)

graph.add_node("calculate_sr", calculate_sr)
graph.add_node("calculate_bpb", calculate_bpb)
graph.add_node("calculate_boundary_percentage", calculate_boundary_percentage)
graph.add_node("summary", summary)


# edges

graph.add_edge(START, "calculate_sr")
graph.add_edge("calculate_sr", "calculate_bpb")
graph.add_edge("calculate_bpb", "calculate_boundary_percentage")
graph.add_edge("calculate_boundary_percentage", "summary")
graph.add_edge("summary", END)
```

```python
workflow = graph.compile()
```

```python
testball_state = {
    "runs": 100,
    "balls": 80,
    "fours": 10,
    "sixes": 5,
    "sr": 0.0,
    "bpb": 0.0,
    "boundary_percentage": 0.0,
    "summary": ""
}
```

```python
workflow.invoke(testball_state)
```
